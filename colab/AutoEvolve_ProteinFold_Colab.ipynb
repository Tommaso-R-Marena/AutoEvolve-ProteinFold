```python
# Configure Git (replace with your details)
import os
from google.colab import userdata

# Get GitHub token from Colab secrets
try:
    GITHUB_TOKEN = userdata.get('GITHUB_TOKEN')
    print('GitHub token loaded from secrets')
except:
    GITHUB_TOKEN = input('Enter your GitHub Personal Access Token: ')

!git config --global user.name "Colab Training Bot"
!git config --global user.email "colab@users.noreply.github.com"

# Configure Git credential helper
!git config --global credential.helper store

# Store credentials
with open('/root/.git-credentials', 'w') as f:
    f.write(f'https://Tommaso-R-Marena:{GITHUB_TOKEN}@github.com')

print('Git configured successfully')
```
